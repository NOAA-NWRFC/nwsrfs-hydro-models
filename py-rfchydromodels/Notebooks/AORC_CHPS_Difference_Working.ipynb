```python
def nse(predictions, targets):
    return 1-np.sum((targets-predictions)**2)/np.sum((targets-targets.mean())**2)

def nnse(predictions, targets):
    return 1/(2-nse(predictions, targets))

def rmse(predictions, targets):
    return np.sqrt(((targets - predictions) ** 2).sum())/np.sqrt(((targets - targets.mean()) ** 2).sum())

def pbias(predictions, targets):
    return 100*(predictions-targets).sum()/targets.sum()

def pbias_norm(predictions, targets):
    return 1-abs(pbias(predictions, targets))/100

def rsquared(predictions, targets):       
    slope, intercept, r_value, p_value, std_err = stats.linregress(targets,predictions)
    return r_value**2

def sdr(predictions, targets):
    return 1-abs(predictions.std()/targets.std()-1)

def f(x):
    d = {}
    d['nse'] = nse(x.sim_flow_cfs, x.flow_cfs)
    d['nnse'] = nnse(x.sim_flow_cfs, x.flow_cfs)
    d['sdr'] = sdr(x.sim_flow_cfs, x.flow_cfs)
    d['pbias'] = pbias(x.sim_flow_cfs, x.flow_cfs)
    d['pbias_norm'] = pbias_norm(x.sim_flow_cfs, x.flow_cfs)
    return pd.Series(d, index=['nse','nnse','sdr' ,'pbias','pbias_norm'])
```

```python
def cyclical_Data(lid):
    sf=pd.read_excel(os.path.join('CHPS_Autocalb_Diff.xlsx'),lid, usecols='A:C',index_col=0,parse_dates=True,engine='openpyxl')
    
    sf['Abs_Diff']=sf.CHPS-sf.AutoCalb
    sf['Perc_Diff']=sf.CHPS/sf.AutoCalb
    sf=sf.resample('1D').mean()
    sf=sf.loc['oct 1980':]
    
    sf_cyclical=sf.groupby(sf.index.dayofyear).mean()
    sf_cyclical['Date']=pd.to_datetime(2099, format='%Y') + pd.to_timedelta(sf_cyclical.index - 1, unit='d')
    #Drop the extra day from the leap day
    sf_cyclical=sf_cyclical[:-1]
    sf_cyclical.reset_index(inplace=True)
    sf_cyclical.loc[sf_cyclical.Date>='10/2099',['Date']]=sf_cyclical.loc[sf_cyclical.Date>='10/2099',['Date']]- pd.DateOffset(years=1)
    sf_cyclical.set_index(['Date'],inplace=True)
    sf_cyclical.sort_index(inplace=True)
    sf_cyclical.drop(['index'],axis=1,inplace=True)
    return sf_cyclical
```

```python
def plot_cyclical_Data(cyclical_data,lid):

    fig, ax = plt.subplots(2,1, dpi= 80, facecolor='w',figsize=(10, 10))

    ax[0].plot(cyclical_data.index,cyclical_data.CHPS,label='CHPS',color='blue')
    ax[0].plot(cyclical_data.index,cyclical_data.AutoCalb,label='Autocalb',color='grey')
    fmt_month = mdates.MonthLocator()
    ax[0].xaxis.set_major_locator(fmt_month)
    ax[0].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
    ax[0].grid(axis='y')
    ax[0].set_ylabel('cfs')

    ax[1].plot(cyclical_data.index,cyclical_data.Perc_Diff, color='red',label='Perc_Diff')
    ax1_secondary = ax[1].twinx()
    ax1_secondary.plot(cyclical_data.index,cyclical_data.Abs_Diff,color='green',label='Total_Diff')
    ax[1].xaxis.set_major_locator(fmt_month)
    ax[1].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
    ax[1].set_ylabel('CHPS/Autocalb')
    ax1_secondary.set_ylabel('CHPS-Autocalb')
    ax[1].grid(axis='x')
    
    fig.suptitle(lid,fontsize=16)
    fig.legend()

    #fig.savefig(os.path.join(folder,'CHPS_AutoCalb_Comparison','Plots',lid+'.png'),bbox_inches = "tight")
    fig.savefig(os.path.join('CHPS_AutoCalb_Comparison',lid+'.png'),bbox_inches = "tight")
    plt.clf()
    plt.close('all')
```

```python
#Format fa_factor dataframe
fa_factors=pd.read_csv(os.path.join(run_dir,'results_01','fa_factors.csv'))
fa_factors['year']=2099
fa_factors['day']=16
fa_factors['hour']=6
fa_factors.index=pd.to_datetime(fa_factors[['year', 'month', 'day','hour']])
fa_factors.drop(['year', 'month', 'day','hour'],axis=1,inplace=True)

#create a dataframe for the manual ptps forcing adjustment and add the original aorc forcings for all zones
manual_ptps_adj=pd.DataFrame(index=pd.date_range(start='9/16/1979',end='11/17/2019',freq='6h'))
for count, df in enumerate(forcing, start=1):
    ptps_series=df.loc[:,['ptps']].squeeze().rename(lid+'-'+str(count))
    manual_ptps_adj=pd.concat([manual_ptps_adj,ptps_series],axis=1)

#merge the fa_factor dataframe into the manual ptps forcing adjustment dataframe at the appropriate month,day,hour 
manual_ptps_adj=manual_ptps_adj.reset_index().merge(fa_factors.ptps_fac, how='outer',left_on=[manual_ptps_adj.index.month,manual_ptps_adj.index.day,manual_ptps_adj.index.hour],
                     right_on=[fa_factors.index.month,fa_factors.index.day,fa_factors.index.hour]).set_index('index').sort_index()
#Drop artifact columns from the merge
manual_ptps_adj.drop(manual_ptps_adj.columns[manual_ptps_adj.columns.str.contains('key')],axis=1,inplace=True)
#Interpolate between the 16th @ 6:00
manual_ptps_adj['ptps_fac']=manual_ptps_adj.ptps_fac.interpolate()
#Remove any indexes outside the POR
manual_ptps_adj=manual_ptps_adj.loc[ac_sacsnow['ptps_fa'].index]

#Create a FA adj column name
fa_columns_names=manual_ptps_adj.columns[manual_ptps_adj.columns.str.contains(lid)]+'_fa'

#Calculate ptps_fa
manual_ptps_adj[fa_columns_names]=manual_ptps_adj.loc[:,manual_ptps_adj.columns.str.contains(lid)].mul(manual_ptps_adj.ptps_fac,axis=0)

#remove any values greater than 1
for fa_col in fa_columns_names:
    manual_ptps_adj.loc[manual_ptps_adj[fa_col]>1,[fa_col]]=1
```
